```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("output.txt", sep="\t")
df
```

```python
plt.figure(figsize=(10, 6), dpi = 120)

plt.title('Performance Comparison', fontsize=20)
plt.xlabel('Array Size', fontsize=14)
plt.ylabel('Time in Second', fontsize=14)

plt.plot(df.Array, df.Bubble, 'b.-', label='Bubble Sort')
plt.plot(df.Array, df.Insertion, 'r.-', label='Insertion Sort')


plt.legend()

plt.savefig('Performance_Comparison.jpg', dpi=300)

plt.show()
```
