```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("All Output.txt", sep="\t")
df
```

```python
plt.figure(figsize=(10, 6), dpi = 120)

plt.title('Performance Comparison', fontsize=20)
plt.xlabel('Array Size', fontsize=14)
plt.ylabel('Time in Second', fontsize=14)

plt.plot(df.Array, df.Insertion, 'b.-', label='Insertion Sort')
plt.plot(df.Array, df.Bubble, 'r.-', label='Bubble Sort')
plt.plot(df.Array, df.Merge, 'g.-', label='Merge Sort')
plt.plot(df.Array, df.Quick, 'y.-', label='Quick Sort')


plt.legend()

plt.savefig('All Performance_Comparison.jpg', dpi=300)

plt.show()
```
