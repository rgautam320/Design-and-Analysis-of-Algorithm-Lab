```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Factorial.txt", sep="\t")
df
```

```python
plt.figure(figsize=(10, 6), dpi = 120)

plt.title('Comparison of Factorial Function using Recursion and Non-Recursion', fontsize=20)
plt.xlabel('Number', fontsize=14)
plt.ylabel('Time in Mili Second', fontsize=14)

plt.plot(df.Number, df.Recursion, 'b.-', label='Using Recursion')
plt.plot(df.Number, df.NonRecursion, 'r.-', label='Using Non-recursion')

plt.legend()

plt.savefig('Factorial Comparison.jpg', dpi=300)

plt.show()
```
